```python
'''
plot 5 sine waves
'''

import numpy as np
import matplotlib.pyplot as plt

col=['red','blue','maroon','purple','green']
for i in range(5):
    #x=np.random.random()
    time=np.arange(100)    #100 is sample rate
    f=np.random.randint(100)    #frequency(cycles per sec)
    amp=np.sin(2*np.pi*f*(time/1000))   #note that 1000 is being used as samples per sec
    plt.plot(time,amp,c=col[i])
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True, which='both')
plt.axhline(y=0,color='k')
plt.show()
```

```python
'''
generate rice generated by all states of India(26 states) and generate the bar and pie chart
'''

states=['AP','Arunachal','Assam',
'Bihar',
'Chhattisgarh',
'Goa','Gujarat',
'Haryana','HP',
'Jharkhand',
'Karnataka','Kerala',
'MP','Maharashtra','Manipur','Meghalaya','Mizoram',
'Nagaland',
'Odisha',
'Punjab',
'Rajasthan',
'Sikkim',
'TN','Telangana','Tripura',
'Uttarakhand','UP',
'WB']

l=len(states)
rice=[]
for i in range(l):
    rice.append(np.random.randint(100000))
xp=np.arange(l)

#bar chart
fig,ax=plt.subplots()
ax.bar(xp,rice,align='center',alpha=0.5)
ax.set_ylabel('Rice in x10^7 kgs')
ax.set_xticks(xp)
ax.set_xticklabels(states)
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()


#pie chart
fig,ax=plt.subplots()
ax.pie(xp,labels=states,autopct='%1.1f%%')
ax.axis('equal')
plt.show()
```

```python
'''
print 3 groups of points on a 2D plane using separate colors
'''

def group():
    n=np.random.randint(100)
    gx=[]
    gy=[]
    for i in range(n):
        gx.append(np.random.randint(10000))
        gy.append(np.random.randint(10000))
    return gx,gy

col=['red','blue','green']
mar=['o','p','s']
for i in range(3):
    x,y=group()
    plt.scatter(x,y,color=col[i],marker=mar[i])
plt.show()
```
