```python
def inp(t):
    if t==0:
        s=input().strip()
        return s
    elif t==1:
        n=int(input().strip())
        return n
    else:
        l=list(map(int,input().split()))
        return l
    
```

```python
#function to find maximum of 3 numbers
def maxthree(a,b,c):
    if a>=b and a>=c: return a
    elif b>=c:return b
    else: return c
```

```python
l=inp(2)
print(maxthree(l[0],l[1],l[2]))
l=inp(2)
print(maxthree(l[0],l[1],l[2]))
l=inp(2)
print(maxthree(l[0],l[1],l[2]))
l=inp(2)
print(maxthree(l[0],l[1],l[2]))
```

```python
#function to find sum of numbers of a list
def sumlist(a):
    s=0
    for i in a:
        s+=i
    return s    
```

```python
l=inp(2)
print(sumlist(l))
```

```python
#function to multiply numbers of a list
def mullist(a):
    s=1
    for i in a:
        s*=i
    return s    
```

```python
l=inp(2)
print(mullist(l))
```

```python
#function to calculate factorial of a number
def factorial(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f    
```

```python
n=inp(1)
print(factorial(n))
n=inp(1)
print(factorial(n))
```

```python
#function to count number of uppercase and lowercase letters
def countcase(s):
    u=0
    l=0
    for i in s:
        x=ord(i)
        if x>=65 and x<=90: u+=1
        elif x>=ord('a') and x<=ord('z'): l+=1
    print("No. of uppercase characters:",u)
    print("No. of lowercase characters:",l)
```

```python
s=inp(0)
countcase(s)
```

```python
#function that takes a list and returns unique elements of the list
def unique(a):
    dic={}
    ans=[]
    for i in a:
        if dic.get(i)==None: 
            ans.append(i)
            dic[i]=True
    return ans       
```

```python
l=inp(2)
print(unique(l))
```

```python
import math as m
#function to check if a number is perfect or not
def perfect(n):
    f=1
    mx=int(m.sqrt(n))+1
    for i in range(2,mx):
        if n%i==0: 
            f+=i
            if n//i!=i: f+=n//i
    return f==n            
```

```python
n=inp(1)
print(perfect(n))
n=inp(1)
print(perfect(n))
n=inp(1)
print(perfect(n))
```

```python
#function to check if a string is palindrome
def palindrome(s):
    l=len(s)
    for i in range(0,l//2+1):
        if s[i]!=s[l-i-1]: return False
    return True    
```

```python
s=inp(0)
print(palindrome(s))
s=inp(0)
print(palindrome(s))
s=inp(0)
print(palindrome(s))
s=inp(0)
print(palindrome(s))
s=inp(0)
print(palindrome(s))
```

```python
'''
function to check if a string is a pangram or not
a pangram is a string which has every alphabet atleast once
'''
def pangram(s):
    c=[False]*26
    for i in s:
        x=ord(i)
        if x>=65 and x<=90: x+=32
        if x>=97 and x<=122: c[x-97]=True
    for i in c:
        if i==False: return i
    return True    
```

```python
s=inp(0)
print(pangram(s))
s=inp(0)
print(pangram(s))
```

```python
'''
function to take a hyphen separated string as input
sort the string in alphabetical order
return the string in the same hyphen separated manner'''
def sorthyphen(s):
    l=s.split("-")
    l.sort()
    res=""
    for i in l:
        if res=="": res+=i
        else: res+="-"+i
    return res        
```

```python
s=inp(0)
print(sorthyphen(s))
s=inp(0)
print(sorthyphen(s))
```

```python
#function to print all permutations of a list
def permutation(a,l,r):
    if l==r: 
        print(a)
        return
    for i in range(l,r+1):
        swap(a,l,i)
        permutation(a,l+1,r)
        swap(a,l,i)

def swap(a,x,y):
    t=a[x]
    a[x]=a[y]
    a[y]=t
    
#function to print all combinations of a list
def combination(a):
    l=len(a)
    m=pow(2,l)
    for i in range(1,m):
        x=[]
        for j in range(0,l):
            if ((i>>j)&1)==1: x.append(a[j])
        print(x)        
```

```python
l=inp(2)
print("Permutations are:")
permutation(l,0,len(l)-1)
print("Combinations are:")
combination(l)
```
